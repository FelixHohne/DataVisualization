```python
import pandas as pd
import collections 
from matplotlib import pyplot as plt

plt.rc('text', usetex=True)

data = pd.read_csv("US_Accidents_May19.csv")
data.drop_duplicates(inplace=True)

ids = data["ID"] 
weather_conditions = data["Weather_Condition"] 
```

```python
counter = collections.Counter()

for data_point in weather_conditions: 
    counter.update([data_point,])
```

```python
weather_conditions = []
frequency = []
for item in counter.most_common(11):
    weather_conditions.append(item[0])
    frequency.append(item[1])

weather_conditions.reverse()
frequency.reverse()

index = 0 

for x in weather_conditions: 
    if str(x) == "nan":
        del weather_conditions[index]
        del frequency[index]
    index +=1

# cleaned_weather_conditions = [x for x in weather_conditions if str(x) != 'nan']
# cleaned_frequency = [x for x in frequency if x != 'nan']
```

```python
plt.style.use("fivethirtyeight")

plt.barh(weather_conditions, frequency)
plt.rc('text', usetex=True)

plt.title("Most Common Weather Conditions during US Traffic Accidents")
plt.xlabel("Number of Accidents")
plt.savefig("weather_conditions.png", dpi=400, bbox_inches='tight')
```
