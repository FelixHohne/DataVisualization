## Top Ten Sources of Remittances from 2000-2016:

From which countries do people send the most remittances? We see that the United States has for the entire period been the largest source of remittances, though its share has been steadily declining. Over time, remittances from other countries such as China and Saudi Arabia now make up a larger share.

**Dataset compiled by:** World Bank

**Source:** https://www.kaggle.com/theworldbank/worldwide-economic-remittances


```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
plt.style.use("fivethirtyeight")
plt.rc('text', usetex=True) # Use Tex formatting
```

```python
df = pd.read_csv("remittance_outflow.csv")
df.transpose()
df.drop([0], axis=0)
df.set_index("Migrant remittance outflows (US$ million)", inplace=True)
df.drop(df.columns[[0]], axis = 1, inplace = True) 
df.drop(df.tail(8).index,inplace=True)
df = df.sort_values("2016", ascending=False)
df = df.head(10)
```

```python
data = df.loc[:, '2000':'2016']
countries = tuple(data.index)
years = np.arange(2000, 2017, 1).tolist()
```

Unique colors for stackplot:

Credit: https://sashat.me/2017/01/11/list-of-20-simple-distinct-colors/

```python
colors = ["#000075", '#3cb44b', '#ffe119', '#4363d8', '#f58231', 
          '#911eb4', '#46f0f0', '#f032e6', '#bcf60c', '#fabebe']
```

```python
plt.figure(figsize=(10.0,6.0))
plt.rc('legend', fontsize=12)   
plt.grid(False)
plt.stackplot(years, data, labels=countries, colors=colors)
plt.ylabel("Millions of USD")
plt.xlabel("Year")
plt.title("Top 10 Sources of Remittances 2000-2016")
plt.xticks(np.arange(2000, 2018, step=2))
plt.tight_layout()
plt.legend(loc="upper left", frameon=False)


plt.savefig("remittances.png", dpi=400)
```
