```python
from matplotlib import pyplot as plt
import pandas as pd
   
plt.rc('text', usetex=True)

plt.style.use("grayscale")
data = pd.read_csv("developer_age.csv")
data.head()
ids = data["Responder_id"]
ages = data["Age"]
bins = [10, 20, 30, 40, 50, 60, 70, 80]
```

```python
# Use log=True to make the histogram logarithmic
plt.hist(ages, bins=bins, edgecolor="white")

median_age = 29
plt.axvline(median_age, color="red", label="Age median", linewidth=2)
plt.title("Programmer Age Distribution", fontsize=20)
plt.xlabel("Ages", fontsize=15)
plt.ylabel("Total Respondents", fontsize=15)
plt.legend(loc="upper right", fontsize=12)

plt.tight_layout()
plt.savefig("developer_salaries.png", dpi=400)
```
